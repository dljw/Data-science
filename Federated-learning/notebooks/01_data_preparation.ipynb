```python
import pandas as pd
import os

from sklearn.model_selection import TimeSeriesSplit
```

```python
df = pd.read_csv("../data/raw/clean_data.csv")
```

```python
df.head()
```

```python
# Splitting data into Stores
store_dir = "../data/store_level_data/"

for store in df['Store'].unique():
    df_sub = df[df['Store'] == store]
    df_sub['Date'] = pd.to_datetime(df_sub['Date'])
    df_sub = df_sub.set_index("Date")
    path = os.path.join(store_dir,"store_" + str(store))
    os.makedirs(path)
    train = df_sub.loc['2012-06-01':].reset_index()
    valid = df_sub.loc['2012-06-02':'2012-10-01'].reset_index()
    test = df_sub.loc['2012-10-01':].reset_index()
    train.to_csv(os.path.join(path, "train.csv"), index=False)
    valid.to_csv(os.path.join(path, "valid.csv"), index=False)
    test.to_csv(os.path.join(path, "test.csv"), index=False)
```
